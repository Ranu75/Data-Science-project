```
def dernierIndiceMaximum(liste):
    maxi = liste[0]
    longueur=len(liste)
    indice_max = 0
    for i in range(longueur):
        if liste[i] >= maxi:
            maxi = liste[i]
            indice_max = i
    return indice_max, maxi
```

```
def word_similarity(word:str):
  D=[]
  word_list = []
  for token in simlex:
    lexeme = nlp.vocab[token.text]
    lexem1 = nlp.vocab[f"{word}"]
    d = lexem1.similarity(lexeme)
    word_list.append(token.text)
    D.append(d)
  position, maxi = dernierIndiceMaximum(D)
  return word_list[position]
```

```
avg_lemme2vec(text,keep=["VERB", "NOUN", ...]):

  1) récupérer les lemmes du texte (enlever ceux qui ne nous intéresse pas) \\
  2) récupérer chaque vecteur de chaque lemme \\
  3) calcul de la moyenne \\
```

```
def get_vector(word:str):
  return nlp.vocab[word].vector
```

```
def avg_lemme2vec(text:str,keep=['ADJ', 'VERB']):
  # on cherche d'abord à récupérer les vecteurs 
  vectors = []
  for token in nlp(text):
#    print(f"working on {token}{token.pos_}")
    if token.pos_ in keep:
#      print('bon token', token)
      # on filtre les token qui ne nous intéressent pas 
      # on recupere le vecteur:
      vectors.append(get_vector(token.lemma_))
  avg_vector = []
  for v in zip(*vectors):
    avg_vector.append(sum(v)/len(v))
  return avg_vector
avg_lemme2vec("There is very interesting. But, I am so tied to understand something. Got to sleep.")
```
